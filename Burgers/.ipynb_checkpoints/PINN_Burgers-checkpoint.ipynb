```python
import numpy as np
import math
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# Set data type
DTYPE=torch.float32

# Set constants
pi=torch.pi
viscosity = .05/pi
#viscosity=0.001

# Define initial condition
def fun_u_0(x):
    return -torch.sin(pi * x)

# Define boundary condition
def fun_u_b(t, x):
    n = x.shape[0]
    return torch.zeros((n,1), dtype=DTYPE)

# Define residual of the PDE
def fun_r(t, x, u, u_t, u_x, u_xx):
    return u_t + u * u_x - viscosity * u_xx
```

```python
# Set number of data points
N_0 = 50
N_b = 50
N_r = 10000

# Set boundary
tmin = 0.
tmax = 1.
xmin = -1.
xmax = 1.



# Lower bounds
lb = torch.tensor([tmin, xmin], dtype=DTYPE)
# Upper bounds
ub = torch.tensor([tmax, xmax], dtype=DTYPE)

# Set random seed for reproducible results
#torch.manual_seed(0)
```

```python
# sample initial value
t_0=torch.ones((N_0,1),dtype=DTYPE)*lb[0]
x_0=torch.rand((N_0,1),dtype=DTYPE)

# scale x_0 to between x_min and x_max
x_0=x_0*(ub[1]-lb[1]) + lb[1]

# concatenate t_0 and x_0 into an N_0 x 2 matrix
X_0 = torch.cat([t_0, x_0], axis=1)


# Evaluate intitial condition at x_0
u_0 = fun_u_0(x_0)

#u_0.shape is torch.Size([50, 1])
```

```python
# Boundary data
t_b = torch.rand((N_b,1), dtype=DTYPE)
t_b = t_b*(ub[0] - lb[0])+lb[0]
x_b = lb[1] + (ub[1] - lb[1]) * torch.bernoulli(torch.ones((N_b,1))-1/2)
# or we can just create two N_b/2 by 1 tensors, one for each boundary


# concatenate to an N_b x 2 matrix
X_b = torch.cat([t_b, x_b], axis=1)


# Evaluate boundary condition at (t_b,x_b)
u_b = fun_u_b(t_b, x_b)



```

```python
# Draw uniformly sampled collocation points
t_r = torch.rand((N_r,1),dtype=DTYPE)
t_r = t_r*(ub[0]-lb[0])+lb[0]
x_r = torch.rand((N_r,1),dtype=DTYPE)
x_r = x_r*(ub[1]-lb[1])+lb[1]
X_r = torch.cat([t_r, x_r], axis=1)



# Collect boundary and inital data in lists
X_data = [X_0, X_b]
u_data = [u_0, u_b]

# X_data[0].shape is torch.Size([50, 2])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(9,6))
plt.scatter(t_0, x_0, c=u_0, marker='X', vmin=-1, vmax=1)
plt.scatter(t_b, x_b, c=u_b, marker='X', vmin=-1, vmax=1)
plt.scatter(t_r, x_r, c='r', marker='.', alpha=0.1)
plt.xlabel('$t$')
plt.ylabel('$x$')

plt.title('Positions of collocation points and boundary data');
#plt.savefig('Xdata_Burgers.pdf', bbox_inches='tight', dpi=300)
```

```python
def init_model(num_hidden_layers=8, num_neurons_per_layer=20):
    layers = []
    layers.append(nn.Linear(2, 20))
    layers.append(nn.Tanh())
    for _ in range(num_hidden_layers):
        layers.append(nn.Linear(20,20))
        layers.append(nn.Tanh())
    layers.append(nn.Linear(20, 1))

    model = nn.Sequential(*layers)
    return model
```

```python
# compute residual in the bulk
# X_r is a (N_r x 2) torch tensor

def get_r(model, X_r):
    
    t, x = X_r[:, 0:1], X_r[:,1:2]
    # this way of slicing makes t and x both (N_r x 1) tensor
    # and NOT a [N_r] vector
    
    t.requires_grad_(requires_grad=True)
    x.requires_grad_(requires_grad=True)
    
    u = model(torch.stack([t[:,0], x[:,0]], axis=1))
    # torch.stack([t[:,0], x[:,0]], axis=1) has shape (N_r x 2)
    # u has shape (N_r x 1)
    
    u_t=torch.autograd.grad(u,t,torch.ones_like(u),create_graph=True)[0]
    u_x=torch.autograd.grad(u,x,torch.ones_like(u),create_graph=True)[0]
    u_xx=torch.autograd.grad(u_x,x,torch.ones_like(u_x),create_graph=True)[0]
    # all derivatives have shape (N_r x 1) because of torch.ones_like(u)
    
    return fun_r(t, x, u, u_t, u_x, u_xx)
```

```python
from torch.utils.data import Dataset

# Data for the bulk collocation points
# no label
# X_r is a (N_r x 2) tensor with features x and t

class FunctionDataset_r(Dataset):
    
    def __init__(self,X_r):
        self.X_r = X_r    
        self.N_r = len(X_r) 
        
    def __getitem__(self,index):
        return self.X_r[index] 
    
    def __len__(self):
        return self.N_r

    
# Data for the initial and boundary points
# with labels
# X_data is a list, X_data[0] and X_data[1] are (50 x 2) tensor
# corresponding to initial and boundary points

class FunctionDataset_b(Dataset):
    
    def __init__(self,X_data,u_data):
        self.X_data = X_data   
        self.u_data = u_data 
        self.N_data = len(X_data)  
        
    def __getitem__(self,index):
        return self.X_data[index],self.u_data[index]
    
    def __len__(self):
        return self.N_data

function_dataset_r = FunctionDataset_r(X_r)
function_dataset_b = FunctionDataset_b(X_data,u_data)

train_dataloader_r = DataLoader(function_dataset_r,batch_size = len(function_dataset_r),shuffle=False)
train_dataloader_b = DataLoader(function_dataset_b,batch_size = len(function_dataset_b),shuffle=False)
```

```python
# Initialize model aka u_\theta
model = init_model()
```

```python
def train(net, epochs, train_dataloader_r,train_dataloader_b,get_r):
    
    
    criterion = nn.MSELoss(reduction = 'mean')
    
    optimizer = optim.Adam(net.parameters(), lr=1e-2)

    # Adam = adaptive momentum optimizer
    # SGD = stastic gradient descent optimizer
    
    
    for epoch in range(epochs): 
        running_loss = 0.0
        if epoch == 1000:
            optimizer = optim.Adam(net.parameters(), lr=1e-3)
        elif epoch == 3000:
            optimizer = optim.Adam(net.parameters(), lr=5e-4)
        elif epoch == 5000:
            optimizer = optim.Adam(net.parameters(), lr=2e-4)
        elif epoch == 7000:
            optimizer = optim.Adam(net.parameters(), lr=1e-4)
        elif epoch == 15000:
            optimizer = optim.Adam(net.parameters(), lr=5e-5)
    
        for i,X_r in enumerate(train_dataloader_r):
            A = get_r(net,X_r)
            loss1 = criterion(A,torch.zeros(A.shape))
        for i,(X_data,u_data) in enumerate(train_dataloader_b):
      
            # get the inputs; data is a list of [inputs, labels]
      
            # zero the parameter gradients
            optimizer.zero_grad()

            # forward + backward + optimize
            outputs = net(X_data)
            loss2 = criterion(outputs,u_data)
            
        loss = loss1 + 15*loss2
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        print('epoch')
        print(epoch)
        print('running_loss')
        print(running_loss)
    print('Finished Training',running_loss, loss2.item())
  
```

```python
train(model, 3000, train_dataloader_r,train_dataloader_b,get_r)
```
